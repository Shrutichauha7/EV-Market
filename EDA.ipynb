```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Setup
sns.set(style='whitegrid')
plt.rcParams["figure.figsize"] = (10, 6)
```

```python
# 2. Load the dataset
df = pd.read_csv("/content/Raw data.zip")
df.head()
```

```python
# 3. Top EV Manufacturers
top_makes = df['Make'].value_counts().head(10)
top_makes.plot(kind='bar', color='skyblue')
plt.title("Top 10 EV Manufacturers by Count")
plt.xlabel("Manufacturer")
plt.ylabel("Number of Vehicles")
plt.xticks(rotation=45)
plt.show()
```

```python

# 4. EV Adoption by Year
year_counts = df['Model Year'].value_counts().sort_index()
sns.lineplot(x=year_counts.index, y=year_counts.values, marker='o')
plt.title("EV Adoption Over Years")
plt.xlabel("Model Year")
plt.ylabel("Number of Vehicles")
plt.show()
```

```python
# 5. Electric Range by Vehicle Type
sns.boxplot(data=df, x='Electric Vehicle Type', y='Electric Range')
plt.title("Electric Range by Vehicle Type")
plt.ylabel("Electric Range (miles)")
plt.xticks(rotation=45)
plt.show()
```

```python
# 6. Top Counties by EV Count
top_counties = df['County'].value_counts().head(10)
top_counties.plot(kind='bar', color='lightgreen')
plt.title("Top 10 Counties by EV Count")
plt.xlabel("County")
plt.ylabel("Number of EVs")
plt.xticks(rotation=45)
plt.show()
```

```python
# 7. CAFV Eligibility Distribution
cafv_counts = df['Clean Alternative Fuel Vehicle (CAFV) Eligibility'].value_counts()
cafv_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title("CAFV Eligibility Distribution")
plt.ylabel("")
plt.show()
```
